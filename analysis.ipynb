```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
df = pd.read_json('assertive_cue0.jsonl', lines=True)
df
```

```python
def categorize_tool_selection(row):
    called_tool = row['called_tool_name']
    
    # No tool called
    if called_tool is None:
        return 'no_tool'
    
    # Check if any of the provided tools is modified and matches the called tool
    for tool in row['tools_provided']:
        if tool['name'] == called_tool:
            if tool.get('modified'):
                return 'modified'
            else:
                return 'non_modified'
    
    # This shouldn't happen if data is consistent
    return 'error'
```

```python

# Apply the function to categorize tool selections
df['tool_selection_category'] = df.apply(categorize_tool_selection, axis=1)

# Calculate statistics
total_calls = len(df)
modified_calls = (df['tool_selection_category'] == 'modified').sum()
non_modified_calls = (df['tool_selection_category'] == 'non_modified').sum()
no_tool_calls = (df['tool_selection_category'] == 'no_tool').sum()

print(f"Total test cases: {total_calls}")
print(f"Modified tool called: {modified_calls} ({modified_calls/total_calls*100:.1f}%)")
print(f"Non-modified tool called: {non_modified_calls} ({non_modified_calls/total_calls*100:.1f}%)")
print(f"No tool called: {no_tool_calls} ({no_tool_calls/total_calls*100:.1f}%)")

# Show breakdown by category
print("\nBreakdown by category:")
print(df['tool_selection_category'].value_counts())

# Display the dataframe
print("\nDataFrame preview:")
print(df[['id', 'called_tool_name', 'tool_selection_category']].head(10))
```
